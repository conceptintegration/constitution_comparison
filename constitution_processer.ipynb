For a document set generate the following model files:

- document_dict: key is an integer document ID, value is a dictionary currently containing `name` key only
- sentence_dict: key is sentence ID in form `<doc_id>/<sentence_number>`, value is sentence text
- sentence_encodings: list of sentence encoding vectors
- encoded_sentences: list of sentence IDs for sentences that pass a selection criterion
