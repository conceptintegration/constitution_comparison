# Constitution Comparison Installer

Installs the the following resources:

- Universal Sentence Encoder (version 4).
- Pre-compiled model files.
- CCP constitutions used to generate the model files.



## Run installation

```python
print("Installing the gdown package...")
!pip install --upgrade --no-cache-dir gdown
print()

# CCP data - Chilean consitutions and Ecuador and Bolivia for comparison
print("Downloading CCP data...")
!gdown 'https://drive.google.com/uc?id=1GiRg1VerJv3pgfl8GIYOk_H6Or5adeaQ' -O data.zip
print("Unzipping data.zip...")
!unzip -o -q 'data.zip'
print("Cleaning up...")
!rm 'data.zip'
!rm -rf '__MACOSX'
print()

# CCP model
print("Downloading CCP model...")
!gdown 'https://drive.google.com/uc?id=1z_H3nzeje8Awhf1p8Rp5ROg7RyFMhNIy' -O model.zip
print("Unzipping model.zip...")
!unzip -o -q 'model.zip'
print("Cleaning up...")
!rm 'model.zip'
!rm -rf '__MACOSX'
print()

# USE model
print("Downloading USE model…")
!gdown 'https://drive.google.com/uc?id=1DgCEP8UMpPfZlZWJpQdm0QKc64p8X_Lz' -O use-4.zip
print("Unzipping use-4.zip…")
!unzip -o -q 'use-4.zip'
print("Cleaning up…")
!rm 'use-4.zip'
!rm -rf '__MACOSX'
print()

print("All downloads complete. You're good to go!")

```
